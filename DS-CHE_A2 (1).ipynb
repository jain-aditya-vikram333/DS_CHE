```python

import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
rng=np.random.RandomState(10)
x=10* rng.rand(50)
y=2*x -5 +rng.randn(50)
plt.scatter(x,y)
```

# Loss Function

```python
def loss(y_p,y):
    
    l1=abs(y_p-y)**3 
    l2=abs(y_p-y)  
    l3=(y_p-y)**4  
    l4=abs(y_p-y)**7
    
    l=l1.mean()+l2.mean()+l3.mean()+l4.mean()
    
    return(l/4)
```

```python
def line(m,x,c):
    return m*x+c
```

```python
m1=np.random.randn(1)
c1=np.random.randn(1)
y_p=line(m1,x,c1)
loss(y_p,y)
```

```python
L=[]
N=x.shape[0]
N
```

# Q2-part 1 -Linear Regression

```python
#liner-A
def grad(y,y_p,N):
    i=0
    dm1=np.zeros(N,dtype="float128")
    
    for i in range(N):
        if(y_p[i]>y[i]):
            dm1[i]=((y_p[i]-y[i])**2)
        else:
            dm1[i]=-(y_p[i]-y[i])**2
        
    
    
    return dm1      
        
```

```python
##linear-A
m1=np.random.randn(1)
c1=np.random.randn(1)
dm=0
dc=0
learningrate=0.0005
numiter=50000

for i in range(numiter):
    
    y_p=line(m1,x,c1)
    ##dm=(3.0/N)*grad(y,y_p,N)
    
    
    dm=(3.0/N)*np.sum(np.multiply(x,grad(y,y_p,N)))
    
    dc=(3.0/N)*np.sum(grad(y,y_p,N))
        
    m1 = m1 -learningrate * dm
    c1 = c1 - learningrate * dc
    if i%5000==0:
        L.append(loss(y_p,y))
        print ("no of iterations:",i )
        plt.plot(x,y,'go')
        plt.plot(x,y_p,'r*')
    #plt.plot(x,m1*x+c1,'r*')
        plt.ylabel('Data')
        plt.show()
    
```

```python
plt.plot(L)
plt.title("LOSS - linear regression-function A ")
```

```python
m1,c1 #linear A
```

# Q2-part 2-Linear Regression 

```python
#liner-B
def grad(y,y_p,N):
    i=0
    dm1=np.zeros(N,dtype="float128")
    
    for i in range(N):
        if(y_p[i]>y[i]):
            dm1[i]=1
        else:
            dm1[i]=-1
        
    
    
    return dm1 
```

```python
##linear-B
m1=np.random.randn(1)
c1=np.random.randn(1)
dm=0
dc=0
learningrate=0.01
numiter=50000

for i in range(numiter):
    
    y_p=line(m1,x,c1)
    ##dm=(3.0/N)*grad(y,y_p,N)
    
    
    dm=(1.0/N)*np.sum(np.multiply(x,grad(y,y_p,N)))
    
    dc=(1.0/N)*np.sum(grad(y,y_p,N))
        
    m1 = m1 -learningrate * dm
    c1 = c1 - learningrate * dc
    if i%5000==0:
        L.append(loss(y_p,y))
        print ("no of iterations:",i )
        plt.plot(x,y,'go')
        plt.plot(x,y_p,'r*')
    #plt.plot(x,m1*x+c1,'r*')
        plt.ylabel('Data')
        plt.show()
```

```python
plt.plot(L)
plt.title("LOSS - linear regression-function B ")
```

```python
m1,c1 #linear B
```

```python
rng=np.random.RandomState(10)
x=10* rng.rand(50)

y=x**2 + x -5 +rng.randn(50)

plt.scatter(x,y)
plt.show()
```

```python
def line(m1,m2,x,c1):
    return m1*x**2+ m2*x +c1
```

```python
m1=np.random.randn(1)
m2=np.random.randn(1)
c1=np.random.randn(1)
y_p=line(m1,m2,x,c1)
plt.plot(x,y,'go')

plt.plot(x,y_p,'r*')
plt.legend(["Given","Predicted"])
plt.show()
```

```python
L=[]
N=x.shape[0]
N
```

```python
m1=np.random.randn(1)
m2=np.random.randn(1)
c1=np.random.randn(1)
dm1=0
dm2=0
dc=0
learningrate=0.00000005
numiter=50000

for i in range(numiter):
    
    y_p=line(m1,m2,x,c1)
    
    grad=(y_p-y)**3
    
    dm1=(4.0/N)*np.sum(np.multiply(x**2,grad))
    dm2=(4.0/N)*np.sum(np.multiply(x,grad))
    
    dc=(4.0/N)*np.sum(grad)
        
    m1 = m1 -learningrate * dm1
    m2=m2 -learningrate * dm2
    c1 = c1 - learningrate * dc
    if i%5000==0:
        L.append(loss(y_p,y))
        print ("no of iterations:",i )
        plt.plot(x,y,'go')
        plt.plot(x,y_p,'r*')
    #plt.plot(x,m1*x+c1,'r*')
        plt.ylabel('Data')
        plt.show()
```
